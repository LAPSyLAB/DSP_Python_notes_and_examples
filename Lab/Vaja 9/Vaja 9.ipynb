## 1. Frekvenčni odziv sistema

Frekvenčni odziv sistema je enak vrednosti z-transforma impulznega odziva $H(z)$, na enotski krožnici.

$H(e^{j\omega})=H(z)|_{z=e^{j\omega}} = \sum_{k=-\infty}^{\infty}h(k)e^{-j \omega k}$

Če je vhodni signal sistema kosinusni signal

$x(n)=A \cos(\omega_0 n + \phi)$,

potem je izhodni signal

$x(n)=A |H(e^{j\omega})|\cos(\omega_0 n + \phi + \theta)$, $\theta = \angle H(e^{j\omega})$.
