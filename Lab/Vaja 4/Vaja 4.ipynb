$y(n) = b(0)\cdot x(n) + b(1)\cdot x(n-1) + ... + b(n_b)\cdot x(n-n_b) - a(1) \cdot y(n-1) - ... - a(n_a)\cdot y(n-n_a)$. 
                  
Velja še $a(0)=1$. Za KEO velja $a(n)=0$ za $n>0$.
