
## 2. Funkcija za prikaz amplitudnega odziva signala

Napišite podprogram, ki bo izračunal in prikazal amplitudni odziv podanega signala. V bistvu je potrebno postopek reševanja 1. naloge zapisati v obliki funkcije, ki sprejme poljubni digitalni signal kot vhodni parameter.

Običajno amplitudne odzive zvočnih signalov prikazujemo v logaritemskem merilu po amplitudni osi – v enotah "decibel" [dB]. Naša percepcija jakosti zvoka je namreč v logaritemski odvisnosti od amplitude zvočnega signala – zato uporaba logaritemske enote pomeni, da signale prikažemo v vizualno enakem razmerju, kot jih slišimo – amplitudni odziv $A(k)=|{X(k)}|$ tako običajno spremenimo še v logaritemsko merilo v dB.

$A_{dB}(k)=20\log_{10}(\frac{A(k)}{A(k)_{max})})$

|40 dB|100|
