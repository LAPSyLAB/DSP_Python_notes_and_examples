## 1. Avtokorelacija 

$r_{xx}(l) = \sum_{n=-\infty}^{\infty} x(n)\cdot x(n-l),\space l=\pm 1,\pm 2,\pm 3,...$

Uporabite posnetka aaa.wav in iii.wav ter posnemite svoj samoglasnik. Primer zajema zvoka:


````python
import sounddevice as sd

trajanje = 1.0   # sekunde
Fs = 44100   	 # Hz
kanali = 1       # mono

# Snemanje zvoka. Zvok je zajet v Numpy array.
print("Recording...")
zvok = sd.rec(int(trajanje * Fs), samplerate=Fs, channels=kanali, dtype='float32', blocking=True)
print("Done...")
````
