Napišite funkcijo ```spectrogram```, ki za posamezne odseke signala (stolpci v signalogramu) izračuna amplitudne odzive in jih zloži v matriko za prikaz amplitudnega kratkočasovnega spektra – t.i. spektrogram. Funkcija ```spektrogram``` sprejme tri parametre: 
- signal
- okno (array s katerim se pomnoži okvir in določa dolžino okvirja)
- overlap (za koliko vzorcev se pomaknemo naprej med zaporednimi okvirji)

index = np.arange(1, n_frames + 1)
