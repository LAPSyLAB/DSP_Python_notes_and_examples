## 1. Z transform, konvolucija

$X(z)=\sum_{n = -\infty}^{\infty} x(n)z^{-n}$

$x*y=Z^{-1}\{X(z)H(z)\}=Z^{-1}\{Z(x(n))Z(h(n))\}$

Preizkusite povezavo med operacijo konvolucije in množenjem polinomov. Izberite dve zaporedji z vsaj petimi neničelnimi, netrivijalnimi koeficient (lahko uporabite zaporedja od prej) in:
- Izračunajte njuno konvolucijo.
- Tvorite Z-transforma obeh signalov (2 polinoma s koeficienti iz obeh zaporedij) in ju pomnožite. Kakšna je povezava z izračunano konvolucijo?

Primer: x=[1 2 3], h=[1 1 1]

````python
import numpy as np

x = np.array([1, 2, 3])
h = np.array([1, 1, 1])
y = np.convolve(x, h)

print(y)
````
